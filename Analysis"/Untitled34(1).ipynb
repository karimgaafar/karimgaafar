```python
#apply inter_quartile_range to handle outliers on numerical values useing numpy but not on original data
data_1=data.copy()


# Define a function to remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply to numeric columns
data_1 = remove_outliers(data_1, "price")
data_1 = remove_outliers(data_1, "value")
data_1 = remove_outliers(data_1, "discount_amount")
data_1 = remove_outliers(data_1, "total")
data_1['price'].head(20)

```

```python
#corrleation represent relationship(positive,negative,non) between age and total sales
#annot display corrleatin values
plt.figure(figsize=(5,4))
column_corrleation=data[['qty_ordered','Total Sales','age']].corr()
sns.heatmap(column_corrleation,annot=True,cmap='coolwarm')
plt.xlabel('Age')
plt.ylabel('Total Sales Spend')
plt.title('corr between Age and Total Sales Spend')
plt.show()
```

```python
#distributon of order complete and this payment method 
#autopct display the percentage on payment method 
plt.figure(figsize=(7, 7))
payment_status=data[data['status']=='complete']['payment_method'].value_counts().head(5)
plt.pie(payment_status,labels=payment_status.index,autopct='%1.1f%%',colors = ['red', 'blue', 'green', 'orange','gray'],startangle=140)
plt.title('top 5 complete orders by payment methods')
plt.show()
```
